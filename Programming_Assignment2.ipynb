```python
#1. Write a Python program to convert kilometers to miles?
#Formula is "miles = km * 0.621371"
km = float(input("Enter distance in kilometers: "))
# conversion factor
miles = km * 0.621371
print(km, "kilometers =", miles, "miles")
```

```python
#2. Write a Python program to convert Celsius to Fahrenheit?
#F = (C Ã— 9/5) + 32
celsius = float(input("Enter the Temp. in Celsius: ") )
farenheight = (celsius* 9/5) + 32 
print(f"This is the temprature in Farenheight: {farenheight}")
```

```python
#3. Write a Python program to display calendar?
import calendar

year = int(input("Enter year: "))
month = int(input("Enter month: "))

print("\n", calendar.month(year, month))
```

```python
#4. Write a Python program to solve quadratic equation?
#shri dharacharya formula laga diya hai yahan pr ROOT = -b+-sqrt(b**2 - 4ac)
#yahan dekhana ye hai ki cmath kon kon si problems solve kr skata hai
import cmath   # complex solutions ke liye

a = float(input("Enter a: "))
b = float(input("Enter b: "))
c = float(input("Enter c: "))

# Discriminant
d = (b ** 2) - (4 * a * c)

# Formula
root1 = (-b + cmath.sqrt(d)) / (2 * a)
root2 = (-b - cmath.sqrt(d)) / (2 * a)

print("Root 1 =", root1)
print("Root 2 =", root2)
```
