```python
#1. Write a Python Program to Find the Factorial of a Number?
num = int(input("Enter a number:"))
fact = 1
for i in range(1,num+1):
    fact = fact*i
print(f"{num}! is {fact}")
```

```python
#2. Write a Python Program to Display the multiplication Table?
Num = int(input("Enter a num to Multiplication Table:"))
for i in range(1,11):
    print(f"{Num} x {i} = {Num*i}")
```

```python
#3. Write a Python Program to Print the Fibonacci sequence?
term = int(input("Enter term how many term you want:"))
a = 0
b = 1
for i in range(1,term+1):
    print(a)
    a,b=b,a+b        
```

```python
 #4. Write a Python Program to Check Armstrong Number?
num = int(input("Enter a number"))
sum = 0
tem = num
digits = len(str(num))
while num>0:
    digit = num % 10
    sum += digit**digits
    num //= 10
    print(num)
if (sum == tem):
    print(tem,"Number is Armstrong")
    
else:
    print(tem,"is not a Armstrong")


```

```python
#5.Write a Python Program to Find Armstrong Number in an Interval?

lowernumber = int(input("Enter the Lower Num."))
uppernumber = int(input("Enter the Upper Num"))
for i in range(lowernumber,uppernumber):
    num = i
    sum = 0
    temp = num
    digits = len(str(num))
    while temp>0:
        digit = temp%10;
        sum += digit**digits
        temp//=10
    if sum == num:
        print(num)
        
    
```

```python
# 6. Write a Python Program to Find the Sum of Natural Numbers?
num = int(input("Enter the num till want to sum. "))
sum=0;
for i in range(1,num+1):
    sum +=i;
print(sum)
```
