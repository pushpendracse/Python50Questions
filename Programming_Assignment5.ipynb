```python
#1. Write a Python Program to Find LCM?
num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))

# Maximum number se start karte hain
if num1 > num2:
    max_num = num1
else:
    max_num = num2

while True:
    if max_num % num1 == 0 and max_num % num2 == 0:
        print("LCM is:", max_num)
        break
    max_num += 1
```

```python
import math

num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))

lcm = (num1 * num2) // math.gcd(num1, num2)#// floor division function and / divide in decimal
print(math.gcd(num1, num2))
print("LCM is:", lcm)
```

```python
#2. Write a Python Program to Find HCF?
num1 = int(input("Enter a number"))
num2 = int(input("Enter another num"))
HCF = 1
if num1>num2:
    minNum = num2
else:
    minNum = num1

for i in range(1,minNum+1):
    if num1 % i==0 and num2 % i == 0:
        HCF = i
print("HCF is : ",HCF)

        
    
    
```

```python
#3. Write a Python Program to Convert Decimal to Binary, Octal and Hexadecimal?

def main():
    try:
        # Taking decimal input from the user
        decimal_number = int(input("Enter a decimal number: "))

        # Converting decimal to binary, octal, and hexadecimal
        binary = bin(decimal_number)        # bin() converts to binary
        octal = oct(decimal_number)          # oct() converts to octal
        hexadecimal = hex(decimal_number)    # hex() converts to hexadecimal

        # Displaying the results
        print("Binary representation:", binary)
        print("Octal representation:", octal)
        print("Hexadecimal representation:", hexadecimal)

    except ValueError:
        # Handling invalid input
        print("Please enter a valid integer.")

# Calling main function
if __name__ == "__main__":
    main()

    
    
```

```python
#4. Write a Python Program To Find ASCII value of a character?
def main():
    try:
        # Taking a character input from the user
        ch = input("Enter a character: ")

        # Checking if only one character is entered
        if len(ch) != 1:
            raise ValueError("Please enter only one character.")

        # Finding ASCII value using ord() function
        ascii_value = ord(ch)

        # Displaying the result
        print("ASCII value of", ch, "is:", ascii_value)

    except ValueError as e:
        # Handling invalid input
        print(e)

# Calling main function
if __name__ == "__main__":
    main()
```

```python
#5. Write a Python Program to Make a Simple Calculator with 4 basic mathematical operations?

def main():
    try:
        # Taking two numbers from the user
        num1 = float(input("Enter first number: "))
        num2 = float(input("Enter second number: "))

        # Displaying operation choices
        print("\nChoose Operation:")
        print("1. Addition (+)")
        print("2. Subtraction (-)")
        print("3. Multiplication (*)")
        print("4. Division (/)")

        # Taking user's choice
        choice = input("Enter your choice (1/2/3/4): ")

        # Performing selected operation
        if choice == '1':
            print("Result:", num1 + num2)

        elif choice == '2':
            print("Result:", num1 - num2)

        elif choice == '3':
            print("Result:", num1 * num2)

        elif choice == '4':
            if num2 != 0:
                print("Result:", num1 / num2)
            else:
                print("Error: Division by zero is not allowed.")

        else:
            print("Invalid choice. Please select 1, 2, 3, or 4.")

    except ValueError:
        # Handling invalid numeric input
        print("Please enter valid numbers.")

# Calling main function
if __name__ == "__main__":
    main()
```
