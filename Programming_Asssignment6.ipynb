```python
#1. Write a Python Program to Display Fibonacci Sequence Using Recursion?
# Recursive function to return nth Fibonacci number
def fibonacci(n):
    if n <= 1:
        return n
    else:
        return fibonacci(n-1) + fibonacci(n-2)

# Main function
def main():
    terms = int(input("Enter number of terms: "))

    if terms <= 0:
        print("Please enter a positive integer")
    else:
        print("Fibonacci sequence:")
        for i in range(terms):
            print(fibonacci(i), end=" ")

# Calling main function
main()

```

```python
#2. Write a Python Program to Find Factorial of Number Using Recursion?
# Recursive function to calculate factorial
def factorial(n):
    if n == 0 or n == 1:     # Base condition
        return 1
    else:
        return n * factorial(n - 1)

# Main function
def main():
    num = int(input("Enter a number: "))

    if num < 0:
        print("Factorial is not defined for negative numbers")
    else:
        result = factorial(num)
        print("Factorial of", num, "is:", result)

# Calling main function
main()

```

```python
#3. Write a Python Program to calculate your Body Mass Index?
def calculate_bmi(weight, height):
    # BMI formula: weight(kg) / height(m)^2
    bmi = weight / (height * height)
    return bmi

def main():
    weight = float(input("Enter your weight in kg: "))
    height = float(input("Enter your height in meters: "))

    bmi = calculate_bmi(weight, height)

    print("Your BMI is:", round(bmi, 2))

    # BMI Category
    if bmi < 18.5:
        print("Category: Underweight")
    elif bmi < 25:
        print("Category: Normal weight")
    elif bmi < 30:
        print("Category: Overweight")
    else:
        print("Category: Obese")


main()

```

```python
#4. Write a Python Program to calculate the natural logarithm of any number?
import math   # math module import

def main():
    num = float(input("Enter a number: "))

    if num <= 0:
        print("Natural logarithm is not defined for zero or negative numbers")
    else:
        result = math.log(num)   # ln(x)
        print("Natural logarithm of", num, "is:", result)


main()
```

```python
#5. Write a Python Program for cube sum of first n natural numbers?

def cube_sum(n):
    sum_cubes = 0
    for i in range(1, n + 1):
        sum_cubes += i ** 3
    return sum_cubes

def main():
    n = int(input("Enter value of n: "))

    if n <= 0:
        print("Please enter a positive integer")
    else:
        result = cube_sum(n)
        print("Cube sum of first", n, "natural numbers is:", result)


main()
```
